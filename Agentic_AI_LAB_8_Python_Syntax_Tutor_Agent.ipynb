```python
#2. Create the Python tutorial function

def get_python_help(topic: str) -> str:
    print("Function called with topic:", topic)

    lessons = {
        "list comprehension": "Creates a list in one line, example: squares = [x*x for x in range(5)].",
        "list": "List stores multiple values, example: nums = [10, 20, 30].",
        "dictionary": "Dictionary stores key-value pairs, example: {'name': 'Ravi'}.",
        "for loop": "For loop repeats code, example: for i in range(3): print(i).",
        "if else": "if-else makes decisions, example: if marks >= 50: print('Pass').",
        "function": "Function is reusable code, example: def greet(): print('Hello')."
    }

    topic = topic.lower()

    for key, value in lessons.items():
        if key in topic:
            return value

    return "Tutorial not available. Available topics: " + ", ".join(lessons.keys())
```

```python
while True:
    user_question = input("Ask a Python syntax question: ")
    answer = get_python_help(user_question)

    print("\nTutor Answer:\n")
    print(answer)
    choice = input("\n Would you like to continue with the next question? Type yes(y) or no(n): ")
    if choice.lower() != "y":
        print("\n Thank you for using the Python Syntax Tutor Agent.")
        break
```
